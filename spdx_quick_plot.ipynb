```python
df = pd.read_csv("~/Downloads/spdx_deleteme2.csv")
df.columns = ["Entity Created Year", "Card ID", "Average amount per card"]
df["Average amount per card"].describe()
df["Average amount per card"] = df["Average amount per card"].apply(lambda x: max(min(x, 2000), 0))

fig, ax = plt.subplots(figsize=(15, 8))
kwargs = {'cumulative': True}
bins = np.arange(0, 2200, 100)
g = sns.distplot(df["Average amount per card"], color="coral", hist_kws=kwargs, kde=True, bins=bins);
g.set_xticks([b for idx, b in enumerate(bins) if idx%2 == 0]);
g.set_yticks([0, 0.25, .5, .75, 1])
#g.set_title("Cumulative distribution of average amount per card");
plt.xlim([0, 2100]);
g.set_xlabel("")
plt.savefig("spdx_sub_analysis_fig2.png")
```
